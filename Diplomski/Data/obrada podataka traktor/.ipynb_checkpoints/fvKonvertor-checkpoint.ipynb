<h3>Sa osciloskopa</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

#Locale settings
import locale
# Set to German locale to get comma decimal separater
locale.setlocale(locale.LC_NUMERIC, "de_DE")

plt.rcParams['text.usetex'] = True 
plt.rcParams.update({'font.size': 10,'font.family': 'serif'})
plt.rcParams['figure.figsize'] = (.75*6.5, .75*4.5)

plt.rcParams['axes.formatter.use_locale'] = True

plt.rcParams['text.latex.preamble'] = r'\usepackage{icomma}\usepackage{upgreek}'
```

```python
freq = pd.read_csv('prenosna/freqData').values / 1000
mean = pd.read_csv('prenosna/meanData').values * 10
```

```python
plt.figure()
plt.plot(freq, mean, 'b')
plt.xlabel('$f$ [kHz]')
plt.ylabel('$v$ [V]')
plt.grid()
plt.savefig('fig_osc/fvPrenosna.pdf', format='pdf', bbox_inches='tight')
```

<h3>Bez filtra</h3>

```python
dataCH2 = pd.read_csv('bezfiltra_tranzistor/CH2csvData').values
```

```python
data = dataCH2
ts = 1/1e3
t = np.linspace(0,ts*len(data),2499)


plt.figure()
plt.plot(t, data*1000, 'b')
plt.xlabel('$t$ [s]')
plt.ylabel('$v$ [mV]')
plt.grid()
plt.savefig('fig_osc/bezFiltra.pdf', format='pdf', bbox_inches='tight')
```

<h3>Sa filtrom na izlazu</h3>

```python
data2CH2 = pd.read_csv('saFiltrom/CH2csvData_genSweep2').values
```

```python
data2 = data2CH2
ts = 1/1e3
t = np.linspace(0,ts*len(data),2499)


plt.figure()
plt.plot(t, data2*1000, 'b')
plt.xlabel('$t$ [s]')
plt.ylabel('$v$ [mV]')
plt.grid()
plt.savefig('fig_osc/saFiltrom.pdf', format='pdf', bbox_inches='tight')
```

```python
plt.figure()
plt.plot(t, data*1000, 'b', label = 'bez filtra na izlazu')
plt.plot(t, data2*1000, 'r', label = 'sa filtrom na izlazu')
plt.xlabel('$t$ [s]')
plt.ylabel('$v$ [mV]')
plt.grid()
plt.legend()
plt.savefig('fig_osc/filtar_preklopljen.pdf', format='pdf', bbox_inches='tight')
```
